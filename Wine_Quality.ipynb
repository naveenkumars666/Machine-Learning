```python
import pandas as pd
```

```python
df1=pd.read_csv("C:/Users/Pravin Kumar/Downloads/winequality-white (1).csv",sep=";")
df1
```

```python
df2=pd.read_csv("C:/Users/Pravin Kumar/Downloads/winequality-red.csv",sep=";")
df2
```

```python
df = pd.concat([df1,df2])
df
```

```python
df['alcohol_buckets'] = pd.qcut(df['alcohol'], q=3, labels=['low', 'medium', 'high'])
df
```

```python
# loop over all columns except 'quality' and create 3 quantile buckets using qcut
for col in df.columns[:-1]:
    df[col+'_buckets'] = pd.qcut(df[col], q=3, labels=['low', 'medium', 'high'])
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# create a list of all bucketed columns
bucket_cols = [col for col in df.columns if '_buckets' in col]

# create a bar chart for each bucketed column
for col in bucket_cols:
    plt.figure()
    sns.countplot(data=df, x=col)
    plt.title(col)
    plt.show()
```
