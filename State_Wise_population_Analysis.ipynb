```python
#Add all the files in the same directory where python is available
import pandas as pd
import glob

path="C:/Users/Pravin Kumar/OneDrive/Desktop/dataset.zip" 
with open('my_file2.csv', 'w') as csv_file:
    #loop to read all txt  files and store in one csv file
    for path in glob.glob('./*.txt',):
        with open(path) as txt_file:
            txt = txt_file.read() + '\n'
            csv_file.write(txt)
df=pd.read_csv("my_file2.csv",header=None, names=["State","Gender","Year","Name","Frequency"])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#plot a graph to know state wise frequency
plt.figure(figsize=(15,7))
plt.scatter(df.State,df.Frequency,alpha=0.6,color='purple')
plt.xlabel('State')
plt.ylabel('Frequency')
plt.title('State VS Frequency ')
```

```python
#creating a crosstab /table of state vs gender
xtab = pd.crosstab(df['State'],df['Gender'])
```

```python
#plot a bar graph for analysis
xtab.plot(kind='bar',stacked=True,figsize=(15,7),color=['Black','orange'],alpha=0.8)
plt.ylabel('Gender')
plt.title("State wise Gender count")
```

```python
plt.figure(figsize=(15,7))
plt.scatter(df.Year,df.Frequency,alpha=0.6,color='Blue')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Year VS Frequency ')
```

```python
#Top 10 Male Names
df_Male = df[df.Gender =='M']
df_Male2 = pd.DataFrame(df_Male.Name.value_counts()[:10]).reset_index()
df_Male2.columns = ['Name', 'Frequency']
df_Male2
```

```python
#Top 10 Female Names
df_Female = df[df.Gender =='F']
df_Female = pd.DataFrame(df_Female.Name.value_counts()[:10]).reset_index()
df_Female.columns = ['Name', 'Frequency']
df_Female
```
